```python
# "r"  --> opens file for reading (default).
# "w"  --> writes new contents while removing the existing ones.
# "x"  --> creates file for writing if it doesnt exists.
# "a"  --> appends content to a existing file.

# "rt" --> text mode.  
# "rb" --> binary mode.

# "r+" --> read and write both. Previous data will be overridden.
# "w+" --> write and read data. override
# "a+" --> append and read data. WONT OVERRIDE  
```

```python
input1a = open('/content/drive/MyDrive/cse221_lab/Trying out I O/input.txt',"rt") 
content = input1a.read()   # if we pass any numerical char in the read() func then it will only read that many particular char.
print(content)

print(input1a.readline(), end = "")    # "readline()" function reads line by line in the text file
print(input1a.readline())
print(input1a.readline())

for line in input1a:    # for traversing line by line in the text file. Condition - line 2 must be commented.
  print(line)

for line in content:    # for traversing char by char
  print(line)

print(input1a.readlines()) #prints everything in a list
input1a.close()
```
